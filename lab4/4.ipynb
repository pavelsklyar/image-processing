```python
def spectr(image):
    imageSpect = np.fft.fft2(image)
    return imageSpect 
```

```python
def centr(image):
    imageSpect = np.fft.fft2(image)
    fshift = np.fft.fftshift(imageSpect)
    return fshift
```

```python
fig = plt.figure()

fig, ([ax1, ax2, ax3 ],[ax4, ax5, ax6] ) = plt.subplots(
    nrows=2, ncols=3,
    figsize=(18, 14)
)

ax1.set_title('1')
ax2.set_title('2')
ax3.set_title('3')
ax4.set_title('4')
ax5.set_title('5')
ax6.set_title('6')

image1 = spectr(image1) 
image2 = spectr(image2)
image3 = spectr(image3)
image4 = spectr(image4)
image5 = spectr(image5)
image6 = spectr(image6)

ax1.imshow(np.log(1+np.abs(image1)), cmap='gray')
ax2.imshow(np.log(1+np.abs(image2)),cmap='gray')
ax3.imshow(np.log(1+np.abs(image3)),cmap='gray')
ax4.imshow(np.log(1+np.abs(image4)),cmap='gray')
ax5.imshow(np.log(1+np.abs(image5)),cmap='gray')
ax6.imshow(np.log(1+np.abs(image6)), cmap='gray')

plt.show()
```
